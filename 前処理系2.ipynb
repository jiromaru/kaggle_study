```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'A' : np.array([15,312,4,24],dtype='float32'),
                   'B' : pd.Timestamp('20130102'),
                   'C' : np.array([1,31,41,124],dtype='float32'),
                   'D' : np.array([1,2,3,4],dtype='int32'),
                   'E' : pd.Categorical(["test","train","test","train"]),
                   'F' : 'foo'})
```

```python
df
```

```python
def standardize_func(df):
    result = df
    for column in df.columns.tolist():
        try:
            mean = df[column].mean()
            std = df[column].std()
            result[column] = (df[column] - mean) / std
        except:
            continue
    return df
```

```python
standardize_func(df)
```
